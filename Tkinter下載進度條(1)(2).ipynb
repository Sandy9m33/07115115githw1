```python
import tkinter as tk
import time

background = tk.Tk() #建立UI畫布
background.title('下載進度條示範') #設置視窗標題名稱
background.geometry('600x180') #設置視窗大小

tk.Label(background, text='下載進度:',).place(x = 20,y = 60)
#設置下載進度條Label(標籤，其中place()是擺放位置，xy座標概念

canvas = tk.Canvas(background, width = 465, height = 22, bg = 'white')
#設置下載進度條的進度Canvas

canvas.place(x = 110, y = 60)
#進度Canvas的位置，xy座標概念

def progress():
    
    #將進度條填滿
    
    fill_line = canvas.create_rectangle(0, 0, 10, 23, width = 0, fill = 'blue')
    
    n=10
    
    for i in range(500):
        n+=1
        canvas.coords(fill_line, (0, 0, n, 23))
        background.update()
        time.sleep(0.01) #調整進度條顯示速度
    
    #將進度條清空
    
    fill_line = canvas.create_rectangle(0, 0, 10, 23, width=0, fill='white')
    
    for i in range(200):
        canvas.coords(fill_line, (0, 0, n, 23))
        background.update()
        time.sleep(0)
        
btn_download = tk.Button(background, text = '開始下載', command = progress)
btn_download.place(x = 400, y = 105)

btn_exit = tk.Button(background, text='離開', command=background.destroy)
btn_exit.place(x = 200, y = 105)

background.mainloop()
```

```python
import tkinter as tk
import requests

window = tk.Tk()

#設置下載進度條

window.title('下載進度條示範 Part2') #設置視窗標題名稱
window.geometry('700x180')

tk.Label(window, text='下載進度:').place(x=40, y=20)
canvas = tk.Canvas(window, width=600, height=16, bg='white')
canvas.place(x=40, y=50)

#下載按鈕函數
def download():
    
    url_str = 'https://statics.teams.cdn.office.net/production-windows-x64/1.4.00.11161/Teams_windows_x64.exe'
    
    response = requests.get(url_str, stream=True)
    #When stream=True is set on the request #保持隨時在request的狀態
    #this avoids reading the content at once into memory for large responses.
    
    chunk_size = 4096 #設置每次下載的塊大小
    
    content_size = int(response.headers['Content-length'])
    #從返回的response的header中獲取文件大小，int轉換成數字型態的資料
    
    #填充進度條
    fill_line = canvas.create_rectangle(1.5, 1.5, 0, 23, width=0, fill='green')
    
    raise_data = 600/ (content_size/chunk_size)
    #進度條每次增加量
    
    global n
    
    #將下載檔案儲存至Jupyter Notebook預設工作路徑
    with open('Teams_windows_x64.exe', 'wb') as f: #僅有檔名產生，但尚未有內涵 #w=write;b=二進位
        
        n=0
        
        for data in response.iter_content(chunk_size=chunk_size): # 每4096Kb寫入一次，逐步確保進度條
            f.write(data) #正式寫入檔案內涵
            n = n + raise_data #遞增h值
            canvas.coords(fill_line, (0, 0, n, 60)) #coords()填滿canvas，第一個參數起點、第二個參數為
            window.update() #讓視窗更新
            
#清空進度條
def clean():
    #清空進度條
    fill_line = canvas.create_rectangle(1.5, 1.5, 0, 23, width=0, fill='white')
    
    for t in range(200):
        canvas.coords(fill_line, (0, 0, n, 60))
        window.update()
        
#下載按鈕

btn_download = tk.Button(window, text = '清空進度條', command = clean)
btn_download.place(x = 130, y = 80)

btn_download2 = tk.Button(window, text = '開始下載', command = download)
btn_download2.place(x = 250, y = 80)

btn_exit = tk.Button(window, text = '離開', command = window.destroy)
btn_exit.place(x = 360, y = 80)

window.mainloop()
```
